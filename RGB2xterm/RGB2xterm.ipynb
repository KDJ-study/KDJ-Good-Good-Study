```python
def hex2rgb ( hexcolor ):
    rgb = [(hexcolor >> 16 ) & 0xff,
            (hexcolor >> 8 ) & 0xff,
           hexcolor  & 0xff
          ]
    return rgb
```

```python
hex2rgb( 0xdadada)
```

```python
kdj_xterm = np.zeros((kdj.shape[0], kdj.shape[1] ))
for w in range(kdj.shape[0]):
    for h in range( kdj.shape[1]):
        R = kdj [w ,h, 0]
        G = kdj [w ,h, 1]
        B = kdj [w ,h, 2]
        min_distance = 65535
        xterm_index = -1
        for i in range(0, len(xterm)):
            RGB = hex2rgb( xterm[i])
            distance = np.abs(R-RGB[0]) + np.abs(G-RGB[1]) + np.abs(B-RGB[2])
            if distance < min_distance:
                xterm_index = i
                min_distance = distance
                kdj_xterm[w,h] = int(xterm_index)
```
